```python
import re
def get_words(file_name):
    with open(file_name, 'r', encoding='utf-8') as f:
       words =  f.read().lower().split()
       words = [re.sub(r'[^a-zA-Z0-9öäå]', '', word) for word in words]
       return words
    
```

```python
import concurrent.futures
import time

def count_words(chunk):
    words = chunk.split()
    return [(word, 1) for word in words]

def parallel_word_count(file_path):
    start_time = time.time()
    with open(file_path, 'r') as file:
        chunks = file.read().lower().split('\n')  # Split the text into chunks for parallel processing
        print(chunks)
        
    with concurrent.futures.ThreadPoolExecutor() as executor:
        word_lists = list(executor.map(count_words, chunks))

    # Flatten the list of lists and create a dictionary to count occurrences
    word_counts = {}
    for word_list in word_lists:
        for word, count in word_list:
            word_counts[word] = word_counts.get(word, 0) + count

    print("--- It took %s seconds to run function ---" % (time.time() - start_time))
    # Convert the dictionary to a list of tuples
    result = list(word_counts.items())
    return result


word_counts_list = parallel_word_count("Text.txt")
print("Word counts:")
for word, count in word_counts_list:
    print(f"{word}: {count}")
```
